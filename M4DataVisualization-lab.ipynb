## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import seaborn as sns 
import matplotlib.pyplot as plt

QUERY = """
SELECT *
FROM master
"""

df = pd.read_sql_query(QUERY,conn)
df.set_index("index", inplace = True )

plt.figure(figsize = (10,8))
sns.histplot(df["ConvertedComp"], bins = 40)
plt.title("Histogram of ConvertedComp")
plt.xlabel("ConvertedComp (USD)")
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
plt.figure(figsize=(10,8))
sns.boxplot( df["Age"])
plt.title('Box Plot of Age')
plt.ylabel('Age')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
plt.figure(figsize=(10, 6)) 
sns.scatterplot(x="Age", y="WorkWeekHrs", data=df)
plt.title("Scatter Plot of Age vs. Work Week Hours") 
plt.xlabel("Age") 
plt.ylabel("Work Week Hours")  
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
plt.figure(figsize=(10,8))
sns.scatterplot(data=df, x='WorkWeekHrs', y='CodeRevHrs', size='Age', alpha = 0.5, legend=True, sizes=(40, 400))
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
QUERY = """
SELECT *
FROM DatabaseDesireNextYear
"""
df_desire = pd.read_sql_query(QUERY,conn)

df_desire.columns = ["Respondent", "Database"]
grouped = df_desire.groupby("Database").size().sort_values(ascending = False)
top_database = grouped.head()

plt.figure(figsize=(10,8))
top_database.plot(kind='pie',
                            autopct='%1.1f%%', 
                            startangle=90,         
                            )
plt.ylabel("Databases")
plt.title("Top 5 Databases Respondents WIsh to Learn Next Year")
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here

new_df = df[["WorkWeekHrs", "CodeRevHrs", "Age"]]

new_df_filtered = new_df[(new_df["Age"]>=30) & (new_df["Age"]<=35)]

grouped_median = new_df_filtered.groupby("Age")[["WorkWeekHrs", "CodeRevHrs"]].median()

grouped_median.plot(kind="bar", stacked=True, figsize=(10, 6))
plt.title("Median Work Week and Code Review Hours for Age Group 30 to 35")
plt.xlabel("Age")
plt.ylabel("Hours")
plt.xticks(rotation=0)  
plt.legend()
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
conv_df = df[["ConvertedComp", "Age"]]
conv_df_filt = conv_df[(conv_df["Age"]>=45) & (conv_df["Age"]<=60)]
groupedMedian = conv_df_filt.groupby("Age")["ConvertedComp"].median()
groupedMedian.plot(kind = "line", x = "Age")
plt.title("Median Converted Compensation for Ages 45 to 60")
plt.xlabel("Age")
plt.ylabel("Median Converted Compensation (USD)")
plt.xlim(44, 62)
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here

mainBranch_df = df["MainBranch"].value_counts()
plt.figure(figsize=(10, 8))  
mainBranch_df.plot(kind="barh")
plt.title("Frequency of Main Branch Categories")
plt.xlabel("Frequency")
plt.ylabel("Main Branch")
plt.show()
```
